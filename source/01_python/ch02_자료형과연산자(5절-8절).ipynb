# 5절. 연산자
## 5.1 산술연산자 : + - * / //(몫연산자), **(제곱), %(나머지)

```python
print('10**5(승) =', 10**5)
print('4/2(결과 실수) =', 4/2)
print('4//2(몫) =', 4//2)
print('4%2(나머지) =', 4%2)
```

```python
# 나누기를 제외한 산술연산은 정수끼리 결과는 정수. 정수와 실수 연산 결과는 실수
print('10**5.0 =', 10**5.0)
print('4*2 =', 4*2)
print('4*2.0 =', 4*2.0)
```

```python
# quiz1. 정수를 입력받아 짝수인지 홀수인지 여부를 출력
num = int(input('정수를 입력하세요 >'))
result = num%2
if result:
    temp = '홀수'
else:
    temp = '짝수'
print("입력하신 수는 {}. 2로 나눈 나머지는 {}, 그래서 {}를 입력하셨군요".format(num, result, temp))
print(f"입력하신 수는 {num}. 2로 나눈 나머지는 {result}, 그래서 {temp}를 입력하셨군요")
```

```python
# quiz2. 정수를 입력받아 5의 배수인지 여부를 출력
num = int(input('정수를 입력하세요 >'))
result = num%5
if result:
    temp = '5의 배수가 아닙니다'
else:
    temp = '5의 배수입니다'
print("입력하신 수는 {}. 5로 나눈 나머지는 {}, 그래서 {}".format(num, result, temp))
print(f"입력하신 수는 {num}. 52로 나눈 나머지는 {result}, 그래서 {temp}")
```

```python
# quiz2. 정수를 입력받아 5의 배수인지 여부를 출력
num = input('정수를 입력하세요>')
if num[-1]=='0' or num[-1]=='5':
    print('5의 배수 입력')
else:
    print('5의 배수 아님')
```
