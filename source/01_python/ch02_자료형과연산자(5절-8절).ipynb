# 5절. 연산자
## 5.1 산술연산자 : + - * / //(몫연산자), **(제곱), %(나머지)

```python
print('10**5(승) =', 10**5)
print('4/2(결과 실수) =', 4/2)
print('4//2(몫) =', 4//2)
print('4%2(나머지) =', 4%2)
```

```python
# 나누기를 제외한 산술연산은 정수끼리 결과는 정수. 정수와 실수 연산 결과는 실수
print('10**5.0 =', 10**5.0)
print('4*2 =', 4*2)
print('4*2.0 =', 4*2.0)
```
