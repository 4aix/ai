# 1절. 조건문(if)
```
if 조건1:
    조건1이 참일 경우 수행할 명령어들(if문 단독 가능. 들여쓰기 중요)
elif 조건2:
    조건1이 거짓이고, 조건2가 참일 경우 수행할 명령어들(elif는 필요한만큼 사용가능)
else:
    모든 조건이 만족하지 않을 때 수행할 명령어들(else는 단독사용불가)
```

````python
score = int(input('Enter your score >'))
if score>=60:
    print("pass입니다")
else:
    print("fail입니다")
````

````python
print("pass입니다" if score>=60 else "fail입니다")
````

````python
score = int(input('Enter your score >'))
# if (score>=90)&(score<=100): 파이썬에서의 비교연산자는 다항연산자
if 90 <= score <= 100:
    print('참 잘했어요')
elif 0<= score < 90:
    print('분발하실거죠?')
else:
    print('유효한 점수가 아닙니다')
````
