# 1절. 조건문(if)
```
if 조건1:
    조건1이 참일 경우 수행할 명령어들(if문 단독 가능. 들여쓰기 중요)
elif 조건2:
    조건1이 거짓이고, 조건2가 참일 경우 수행할 명령어들(elif는 필요한만큼 사용가능)
else:
    모든 조건이 만족하지 않을 때 수행할 명령어들(else는 단독사용불가)
```

````python
score = int(input('Enter your score >'))
if score>=60:
    print("pass입니다")
else:
    print("fail입니다")
````

````python
print("pass입니다" if score>=60 else "fail입니다")
````

````python
score = int(input('Enter your score >'))
# if (score>=90)&(score<=100): 파이썬에서의 비교연산자는 다항연산자
if 90 <= score <= 100:
    print('참 잘했어요')
elif 0<= score < 90:
    print('분발하실거죠?')
else:
    print('유효한 점수가 아닙니다')
````

````python
print('참 잘했어요' if 90<=score<=100 else \
      '분발하실거죠?' if 0<=score<90 else \
      '유효한 점수가 아닙니다')
````

````python
# 점수를 입력받아 학점(A, B, C, D, F, 유효하지 않은 점수)
# 결과 : 입력하신 점수는 70점이고 C등급입니다 (if문, if식)
score = int(input('Enter your score >'))
if 90<=score<=100:
    grade = 'A'
elif 80<=score<90:
    grade = 'B'
elif 70<=score<80:
    grade = 'C'
elif 60<=score<70:
    grade = 'D'
elif 0<=score<60:
    grade = 'F'
else:
    grade = '유효하지 않은 점수'
print("입력하신 점수는 {}점이고 {}등급입니다".format(score, grade))
````

````python
grade = 'A' if 90<=score<=100 else \
        'B' if 80<=score<90 else \
        'C' if 70<=score<80 else \
        'D' if 60<=score<70 else \
        'F' if 0<=score<60 else \
        '유효하지 않은 점수'
print("입력하신 점수는 {}점이고 {}등급입니다".format(score, grade))
````

# 2절. 반복문
## 2.1 for-in문
```
for 임시변수 in 나열가능한자료:
    반복할 문장들
else:
    나열 가능한 자료를 다 소요하여 for블록을 다 수행한 후, 최종 한번 실행할 문장들
    for문안의 break를 수행할 경우, else블록을 수행하지 않음(else는 선택)
```
