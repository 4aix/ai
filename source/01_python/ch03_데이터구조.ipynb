```python
# listмқҳ itemл“Ө м¶ңл Ҙн•ҳкё°
fruits = ['mango', 'apple', 'orange', 'grape']
for idx in range(len(fruits)): # [0,1,2,3]  len(fruits):fruitsмқҳ item к°ҜмҲҳ
    print("{}лІҲм§ё кіјмқјмқҖ {}".format(idx, fruits[idx]))
#     print(idx, 'лІҲм§ё кіјмқјмқҖ', fruits[idx])
```

```python
list(enumerate(fruits))
```

```python
for idx, value in enumerate(fruits):
    print("{}лІҲм§ё кіјмқјмқҖ {}".format(idx, value))
```

```python
for item in fruits:
    print(item)
```

## 1.2 кё°ліё м •ліҙ мЎ°нҡҢ

```python
numbers = [1, 12, 3, 4, 5]
numbers_2d = [[100, 0, 1, 2, 3],
              [9, 100, 110, 102, 103],
              [20, 30, 40]]
```

```python
# мҡ”мҶҢмҲҳ
print(len(numbers), len(numbers_2d), len(numbers_2d[1]))
```

```python
# мөңмҶҢк°’, мөңлҢҖк°’, н•©кі„
min(numbers), max(numbers), sum(numbers)
```

```python
# 2м°Ёмӣҗ listмқҳ min(): л¶Җл¶„ лҰ¬мҠӨнҠёмқҳ мІ«мҡ”мҶҢк°Җ к°ҖмһҘ мһ‘мқҖ л¶Җл¶„лҰ¬мҠӨнҠё
#             max(): л¶Җл¶„ лҰ¬мҠӨнҠёмқҳ мІ«мҡ”мҶҢк°Җ к°ҖмһҘ нҒ° л¶Җл¶„лҰ¬мҠӨ
min(numbers_2d), max(numbers_2d)
```

```python
fruits = ['mango', 'banana', 'apple']
min(fruits), max(fruits)
```

```python
# м •л ¬н•ҳкё°
numbers = [1, 12, 3, 4, 5]
print('мҳӨлҰ„м°ЁмҲң м •л ¬ :', sorted(numbers))
print('лӮҙлҰјм°ЁмҲң м •л ¬ :', sorted(numbers, reverse=True))
print('sortedн•ЁмҲҳ мӮ¬мҡ©мӢң лҰ¬мҠӨнҠё лӮҙмҡ©мқҖ л°”лҖҢм§Җ м•ҠмқҢ :', numbers)
```
