## 1.4 인덱싱과 요소 인덱스 찾기
- 변수[idx] : 인덱스를 이용한 접근(없는 인덱스 사용시 에러)
- count(찾을데이터) : 해당 찾을 데이터가 나오는 갯수 반환
- index(찾을데이터) : 0번째부터 해당 찾을데이터가 몇번째 index에 있는지 반환
- index(찾을데이터, start) : start번째부터 해당 찾을데이터가 몇번째 index에 있는지 반환

```python
numbers = [1, 2, 3, [1, 2, 3], 1, 2, 3]
```

```python
numbers.count(2)
```

```python
numbers.count(9)
```

```python
print(2 in numbers) # 2가 numbers의 요소인지 여부
print(9 in numbers)
```

```python
numbers.index(2, 2)
```

```python
numbers.index(9)
```

```python
if numbers.count(9):
    print(numbers.index(9))
else:
    print('찾을 수 없음')
```

```python
if 9 in numbers:
    print(numbers.index(9))
else:
    print('찾을 수 없음')
```

```python
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 반환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in lst:
        return lst.index(item)
    return -1
```
