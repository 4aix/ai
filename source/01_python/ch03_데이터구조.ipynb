## 1.1 리스트 만들기
    - []를 이용해서 만들며, 여러개 값(여러 타입)을 하나의 변수에 저장 관리할 수 있음
    - list()함수 이용도 가능
    - 인덱스 : 각요소의 위치, 인덱스는 0부터 시작. 맨마지막 -1
    - 슬라이싱([from:stop:step]) : 부분 데이터 추출
    - 인덱스와 슬라이싱을 이용해서 쓰기와 읽기 지원

```python
fruits = ['banana','apple','orange','grape']
type(fruits)
```

```python
print(fruits)
print(fruits[0])
```

```python
fruits[0] = 'mango' # 0번째 item 수정
print(fruits)
```

```python
numbers = [10, 20]
a, b = numbers
print(a, b)
```

```python
# range(from, to, by) : from부터 to전까지 by씩 증가(감소)하면서 지정된 범위의 나열가능한 자료
        # from 생략시 0
        # by 생략시 1
range(3)
```

```python
list(range(3))
```

```python
list(range(20, 10, -2))
```

```python
for i in range(3):
    print(i, '안녕하세요')
```

```python
# list의 item들 출력하기
fruits = ['mango', 'apple', 'orange', 'grape']
for idx in range(len(fruits)): # [0,1,2,3]  len(fruits):fruits의 item 갯수
    print("{}번째 과일은 {}".format(idx, fruits[idx]))
#     print(idx, '번째 과일은', fruits[idx])
```

```python
list(enumerate(fruits))
```

```python
for idx, value in enumerate(fruits):
    print("{}번째 과일은 {}".format(idx, value))
```

```python
for item in fruits:
    print(item)
```

```python
# 리스트는 여러 타입의 데이터를 담을 수 있음
numbers = [0, True, '2', 3, [4, 5, 6]]
print(numbers)
print(numbers[4][0])
```

```python
# 2차원 리스트
numbers_2d = [[1, 2, 3, 4, 5],
              [11, 12, 13, 14, 15],
              [21, 22, 23, 24, 25]]
len(numbers_2d), len(numbers_2d[0])
```

```python
numbers_2d[1][1]
```

```python
for i in range(len(numbers_2d)):
    for j in range(len(numbers_2d[i])):
        print(f"{i}행{j}열 : {numbers_2d[i][j]}", end='\t')
    print() # 개행
```

## 1.2 기본 정보 조회

```python
numbers = [1, 12, 3, 4, 5]
numbers_2d = [[100, 0, 1, 2, 3],
              [9, 100, 110, 102, 103],
              [20, 30, 40]]
```

```python
# 요소수
print(len(numbers), len(numbers_2d), len(numbers_2d[1]))
```

```python
# 최소값, 최대값, 합계
min(numbers), max(numbers), sum(numbers)
```

```python
# 2차원 list의 min(): 부분 리스트의 첫요소가 가장 작은 부분리스트
#             max(): 부분 리스트의 첫요소가 가장 큰 부분리스
min(numbers_2d), max(numbers_2d)
```

```python
fruits = ['mango', 'banana', 'apple']
min(fruits), max(fruits)
```

```python
# 정렬하기
numbers = [1, 12, 3, 4, 5]
print('오름차순 정렬 :', sorted(numbers))
print('내림차순 정렬 :', sorted(numbers, reverse=True))
print('sorted함수 사용시 리스트 내용은 바뀌지 않음 :', numbers)
```

## 1.3 요소 추가

```python
# + : 두 리스트 연결
# * : 리스트를 곱한 수만큼 반복
# append() : 요소 맨 뒤에 추가
# extend() : 리스트를 요소별로 맨 뒤에 추가
# insert(idx, 값) : idx번째 인덱스 위치에 값 추가
```

```python
numbers = [1,2,3]
numbers = numbers + numbers # 리스트 연결
numbers
```

```python
numbers = [1,2,3]
print(numbers * 3)
print(3 * numbers)
```
