# 1절. 함수의 정의 및 사용
## 1.1 함수정의하기

````python
# 함수정의
a = 10
def my_message(n, msg): # msg를 n번 출력하는 함수
    for i in range(n):
        print(i, msg)
````

````python
# 함수 사용 == 함수 호출
my_message(5, 'hello')
my_message(2, '홍길동')
````
