```python
# 함수정의
def my_message(n, msg): # msg를 n번 출력하는 함수
    for i in range(n):
        print(i, msg)
```

```python
# 함수 사용 == 함수 호출
my_message(5, 'hello')
my_message(2, '홍길동')
```
