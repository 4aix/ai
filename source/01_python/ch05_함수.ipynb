## 1.1 함수정의하기

````python
# 함수정의
a = 10
def my_message(n, msg): # msg를 n번 출력하는 함수
    for i in range(n):
        print(i, msg)
````

````python
# 함수 사용 == 함수 호출
my_message(5, 'hello')
my_message(2, '홍길동')
````

````python
# 함수 중복 정의 불가(오버로딩 불가)
def my_message(n, msg): # msg를 n번 출력하는 함수
    for i in range(n):
        print(i, msg)
def my_message(cnt:int): # python을 cnt번 출력
    for i in range(cnt):
        print('Hello, Python', end='\t')
        print('Hi, Python')
````
