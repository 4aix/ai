## 1.3 지역변수(local var)와 전역변수(global var)

````python
global_var = 100 # 전역변수
def func1():
    print('함수안 :', global_var)
func1()
print('함수밖 :', global_var)
````

````python
def func2():
    local_var = 200 # 지역변수
    print('함수안 :', local_var)
func2()
print('함수밖 :', local_var)
````

````python
# 변수의 값을 조회하는 순서 : 지역변수(지역심볼테이블) > 전역변수(전역심볼테이블) > 내장된 이름
g_var = 100 # 전역변수
def func3():
    g_var = 200 # 지역변수
    print('함수안 g_var :', g_var)
func3()
print('함수밖 g_var :', g_var)
````

````python
# 렉시컬 특징 : 변수가 어디에 선언(할당)되었는지에 따라 변수 범위가 정해짐
g_var = 100 # 전역변수
def func3():
    print('함수안 g_var :', g_var)
    g_var = 200 # 지역변수
func3()
print('함수밖 g_var :', g_var)
````

````python
# 렉시컬 특징 : 변수가 어디에 선언(할당)되었는지에 따라 변수 범위가 정해짐
g_var = 100 # 전역변수
def func3():
    global g_var # func3함수내의 g_var는 전역변수를 쓴다
    print('함수안 g_var :', g_var) # 전역변수
    g_var = 200 # 전역변수
func3()
print('함수밖 g_var :', g_var)
````

## 여러개 값을 return

````python
def swap(a, b):
    "매개변수의 순서를 바꿔서 return"
    return b, a # 여러개 값을 return하면 tuple로 전달
````

## 3.3 순서인수, 키워드인수
- 순서인수 : 함수의 정의된 순서대로 전달방식
- 키워드인수 : 인수의 이름을 명시하는 방식

````python
def func(a, L=None):
    if L is None:
        L = []
    L.append(a)
    return L
````

````python
lst = []
print(func(10, lst)) # 순서인수
print(func(L=[], a=10)) #키워드인수
print(func(20)) # 순서인수 a에 20이 전달
print(func(20, None))
````

## 3.4 튜플 매개변수를 이용한 가변인수 설정

````python
# 함수 정의
def add(a, b):
    return a+b
def add(a, b, c):
    return a+b+c
````

````python
print(add(10, 20))
print(add(10, 2, 3))
````

````python
# args앞의 * : 매개변수들(인수들)이 args 튜플로 전달
def add(*args):
    "전달받은 매개변수들의 합을 return. 매개변수가 없을 경우 0을 return"
    #print(args) # ()
    tot = 0 # 누적변수
    for num in args:
        tot += num # tot = tot + num
    return tot
````

````python
print(add())
print(add(1))
print(add(10, 20))
print(add(10, 2, 3))
print(add(1, 2, 3, 4, 5, 3, 4, 5, 6, 7, 8, 9))
````

````python
subject = ('python', 'web', 'Oracle', 'MySQL', 'DL', 'ML', 'LLM', 'flask/django')
" - ".join(subject)
````

````python
# 튜플 매개변수와 일반 매개변수가 같이 있을 때
def concat(*subject, sep):
    return sep.join(subject)
````

## 3.5 딕셔너리 매개변수
- 여러 개의 키워드인수를 딕셔너리 형태로 받아서 가변 키워드 인수로 처리

````python
def func(**args):
    print(args)
    print(type(args))
    for key, value in args.items():
        print("{}:{}".format(key, value))
    print("color :", args.get('color', 'black'))
    print('width :', args.get('width', 1))
    print('point :', args['point'])
````

````python
func(color="red", width=2, point="O", name='선그래프')
````

## 3.6 함수 정의시 매개변수의 순서
- 순서인수 > 튜플인수(*) > 키워드인수(기본값) > 딕셔너리인수(**)

````python
def func1(a, b, *c, **d):
    print("순서인수 a =", a)
    print("순서인수 b =", b)
    print("튜플인수 c =", c)
    print("dict인수 d =", d)
````

````python
func1('a', 'b', 10,20,30,40,50,60, name="홍길동", age=30, add="서울")
````

````python
func1(10, 20)
````

````python
# 순서인수 > 튜플인수(*) > 키워드인수(기본값) > 딕셔너리인수(**)
def func2(a, b, c, *d, e=10, **f):
    print("순서인수 a =", a)
    print("순서인수 b =", b)
    print("순서인수 c =", c)
    print("튜플인수 d =", d)
    print("키워드인수 e =", e)
    print("딕셔너리인수 f =", f)
````
