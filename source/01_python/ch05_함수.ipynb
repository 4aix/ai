## 1.3 지역변수(local var)와 전역변수(global var)

````python
global_var = 100 # 전역변수
def func1():
    print('함수안 :', global_var)
func1()
print('함수밖 :', global_var)
````

````python
def func2():
    local_var = 200 # 지역변수
    print('함수안 :', local_var)
func2()
print('함수밖 :', local_var)
````

````python
# 변수의 값을 조회하는 순서 : 지역변수(지역심볼테이블) > 전역변수(전역심볼테이블) > 내장된 이름
g_var = 100 # 전역변수
def func3():
    g_var = 200 # 지역변수
    print('함수안 g_var :', g_var)
func3()
print('함수밖 g_var :', g_var)
````

````python
# 렉시컬 특징 : 변수가 어디에 선언(할당)되었는지에 따라 변수 범위가 정해짐
g_var = 100 # 전역변수
def func3():
    print('함수안 g_var :', g_var)
    g_var = 200 # 지역변수
func3()
print('함수밖 g_var :', g_var)
````

````python
# 렉시컬 특징 : 변수가 어디에 선언(할당)되었는지에 따라 변수 범위가 정해짐
g_var = 100 # 전역변수
def func3():
    global g_var # func3함수내의 g_var는 전역변수를 쓴다
    print('함수안 g_var :', g_var) # 전역변수
    g_var = 200 # 전역변수
func3()
print('함수밖 g_var :', g_var)
````

## 1.4 값에 의한 호출
- 함수의 인수로 전달되는 변수가 스칼라변수(문자, 숫자, 논리형)일 때

````python
foo = 'a' # 전역변수
id(foo)
````

````python
def func1(f): # 매개변수 f는 지역변수
    print("before :", f, '의 주소는', id(f))
    f = f*3 # 'aaa'
    print("after :", f, '의 주소는', id(f))
func1(foo)
print('foo :', foo, '의 주소는', id(foo))
````

## 1.5 참조에 의한 호출
- 함수의 인수로 전달되는 변수가 리스트, 튜플, 셋, 딕셔너리, 객체일 때

````python
lst = [1, 2, 3] # 전역변수
id(lst)
````

````python
def func1(f): # 매개변수 f는 지역변수
    print("before :", f, '의 주소는', id(f))
    f.append(4)
    print("after :", f, '의 주소는', id(f))
func1(lst)
print('lst :', lst, '의 주소는', id(lst))
````

````python
func1(lst)
print(lst)
````
