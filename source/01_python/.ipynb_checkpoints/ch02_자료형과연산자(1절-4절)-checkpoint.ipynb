# 1절. 변수
(변수 : 프로그램 작업 처리를 위해, 값을 저장할 수 있는 메모리 공간)
- 파이썬은 변수 선언을 위한 타입이 없고, 별도의 선언 과정도 없음

## 1.1 변수 할당
- 할당연산자(=)를 이용해서 값을 할당하면 타입이 셋팅
- 파이썬에서의 모든 변수는 객체로 저장

```python
r = 5
area = 3.14 * r * r
print('계산된 넓이는', area)
```

```python
import numpy as np # numpy라는 패키지(폴더)를 np라는 이름으로 사용
```

```python
area = np.pi * r * r
```

```python
r = 4
print('계산된 넓이는', area)
```

```python
print('area주소 :', id(area))
print('area타입 :', type(area))
```

```python
area = 10
print('area주소 :', id(area))
print('area타입 :', type(area))
```

```python
area = "파이썬"
print('area주소 :', id(area))
print('area타입 :', type(area))
```

## 1.2 변수이름 규칙
- 문자(특수문자는 _를 제외하고는 안됨)로 시작
- 문자, 숫자, _로 변수명 작성
- 함수명은 변수명으로 가능하나 피한다. 꼭 써야한다면 str_식으로 _를 붙여 사용
- 파이썬 예약어(for, if, ...)는 변수이름 사용 불가
- 한글변수는 꼭 필요시에만 사용

```python
이름 = "홍길동" # 비추
print(이름)
```

```python
for = 10
```

```python
print = 10
print(1)
```

```python
del print # print 변수 삭제
print(1)
```
