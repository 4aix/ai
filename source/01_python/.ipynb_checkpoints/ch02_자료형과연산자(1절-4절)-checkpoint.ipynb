# 1절. 변수
(변수 : 프로그램 작업 처리를 위해, 값을 저장할 수 있는 메모리 공간)
- 파이썬은 변수 선언을 위한 타입이 없고, 별도의 선언 과정도 없음

## 1.1 변수 할당
- 할당연산자(=)를 이용해서 값을 할당하면 타입이 셋팅
- 파이썬에서의 모든 변수는 객체로 저장

```python
r = 5
area = 3.14 * r * r
print('계산된 넓이는', area)
```

```python
import numpy as np # numpy라는 패키지(폴더)를 np라는 이름으로 사용
```

```python
area = np.pi * r * r
```

```python
r = 4
print('계산된 넓이는', area)
```

```python
print('area주소 :', id(area))
print('area타입 :', type(area))
```

```python
area = 10
print('area주소 :', id(area))
print('area타입 :', type(area))
```

```python
area = "파이썬"
print('area주소 :', id(area))
print('area타입 :', type(area))
```

```python
def add(x, y=0):
    '''
    매개변수 2개를 받아 두자리 숫자로 만들어요
    x = 10의 자리수가 됨
    y = 1의 자리수가 됨(default 값 0)
    사용법 : add(2, 3) => 23
    '''
    return x*10 + y
```

```python
add(5, 4)
```

```python
add(5)
```

```python
a = add
a(5, 3)
```

## 1.4 다중 변수 선언

```python
a = 10
b = 20
```

```python
a, b = 10, 20 # 동시에 다중 변수 선언과 할당
```

```python
print(a, b)
```

```python
a, b = b, a+b # 다중 변수 할당시 순차적 할당이 아니고 일괄 동시 할당
```

```python
a, b
```
